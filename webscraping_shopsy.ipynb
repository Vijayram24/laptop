```python
import pandas as pd
import requests 
from bs4 import BeautifulSoup
n=[]
p=[]
for i in range(29):
    url=f'https://www.shopsy.in/search?q=laptop&page={i+1}'
    r = requests.get(url)
    
    soup = BeautifulSoup(r.text , 'lxml')

    names = soup.find_all('div' , class_ = 'sc-46489703-1 bjxoUG')
    for name in names:
        desc=name.find('div' , class_ = 'sc-c50e187b-0 bkNEtl')
        price=name.find('div' , class_ = 'css-146c3p1 r-cqee49 r-1vgyyaa r-ubezar r-1rsjblm r-13hce6t')
        
        try :
            print(desc.get_text(),"  ",price.get_text())
            n.append(desc.get_text())
            p.append(price.get_text())
        except :
            pass
```

```python
data={}

data['name']=n
data['price']=p
df=pd.DataFrame(data)
df.shape
```

```python
df.to_csv('laptop_project.csv')
```
